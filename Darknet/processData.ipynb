```python
import os
import cv2
import matplotlib.pyplot as plt
from tqdm import tqdm
import shutil
FOLDERPATH = './test'
FOLDERPATH = os.path.abspath(FOLDERPATH)
OUTPUTPATH = './result'
OUTPUTPATH = os.path.abspath(OUTPUTPATH)
if os.path.exists(OUTPUTPATH):
    shutil.rmtree(OUTPUTPATH)  
    os.makedirs(OUTPUTPATH)
else:
    os.makedirs(OUTPUTPATH)
folderlist = os.listdir(FOLDERPATH)
```

```python
print(os.getcwd())
print(os.path.abspath('.'))
```

```python
num = 0
for folder in folderlist:
    folderPath = os.path.join(os.path.abspath(FOLDERPATH), folder)
    for i in tqdm(os.listdir(folderPath)):
        #os.rename(os.path.join(folderPath,i), folderPath+"/"+str(num)+"_"+folder+"."+os.path.basename(i).split(".")[-1])
        newPath = OUTPUTPATH+"/"+str(num)+"_"+folder+"."+os.path.basename(i).split(".")[-1]
        shutil.copyfile(os.path.join(folderPath,i), newPath)
        num += 1
```
